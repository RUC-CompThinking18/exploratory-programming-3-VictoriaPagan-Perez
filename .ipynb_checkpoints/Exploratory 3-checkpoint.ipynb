```python
import re#'Import re' is what causes Python to recognize genex code.
source = open("text.txt")  #I am telling the program that I want 'text.txt' to be analyzed. 
book = source.read()
def count(book):      
    if type(book) != str :# I told the program that if the 'book' is not a string, to raise a type error with the text 'This is not a string'.
        raise TypeError('This is not a string!')  #This TypeError not only raises itself, but it also cuts the rest of the program, since it didn't meet the parameters I set.
    word_bunch = re.findall(r"\w*at\b", book)#I am telling python to call on genex codes and look for the parameters given (red text)[variable 1] in the second variable 'book'.
    four_letters = []  #I am creating an empty string to add all of the words that meet the requirements below. 
    for word in word_bunch:
        if len(word) > 3:
            four_letters.append(word) #I am telling the program to remove any of the words that are above three characters in length and to add them into the 'four_letters' string. 
     #This gives me the list that I want and I then tell the program to return the answer.
    return four_letters
            
    
```

```python
count(book)
```
