```python
import re
source = open("text.txt")
book = source.read()
def count(book):
    if type(book) != str :
        raise TypeError('This is not a string!')
    word_bunch = re.findall(r"\w*at\b", book)
    four_letters = []
    for word in word_bunch:
        if len(word) > 3:
            four_letters.append(word)
            
    return four_letters
            
    
```

```python
count(book)
```
